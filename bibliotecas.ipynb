```python
#Importação de bibliotecas
import numpy as np #Será o apelido
```

```python
temperaturas = np.array([1000, 40, 50, 60, 70, 90, 100])
print(temperaturas)
print(type(temperaturas))
```

```python
#Média Aritmética
media = np.mean(temperaturas)
print(f"{media:.2f}")
```

```python
#Retorna o maior valor da lista
maior_valor = np.max(temperaturas)
print(maior_valor)
```

```python
#Retornar o menor valor da lista
menor_valor = np.min(temperaturas)
print(menor_valor)
```

```python
#Salários x 1000
salarios = np.array([2, 3, 4, 5, 1, 500 ])
media_salarios = np.mean(salarios)
print(f"{media_salarios:.2f}")
```

```python
mediana = np.median(salarios)
print(mediana)
```

Desafio final:
- Criar um array de habitantes de 5 cidades do estado de SP.
- Maior população.
- Menor população.
- Média de população.
- Mediana

```python
habitantes = np.array([247.500, 189.456, 301.292, 64.596, 291.116])
maior_pop = np.max(habitantes)
menor_pop = np.min(habitantes)
media_pop = np.mean(habitantes)
mediana_pop = np.median(habitantes)

print(f"Americana: {habitantes[0]:.3f} habitantes\nSanta Bárbara d'Oeste: {habitantes[1]} habitantes\nLimeira: {habitantes[2]} habitantes\nNova Odessa: {habitantes[3]} habitantes\nSumaré: {habitantes[4]} habitantes")
print(f"\nMaior População: {maior_pop}")
print(f"Menor População: {menor_pop}")
print(f"População Média: {media_pop}")
print(f"Mediana: {mediana_pop}")
```
