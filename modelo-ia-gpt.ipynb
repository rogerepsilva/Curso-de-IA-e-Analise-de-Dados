## N-Gramas (n-grams)

```python
frase = "Águias observam montanhas enquanto cientistas desenvolvem algoritmos robustos inspirando comunidades curiosas mediante colaboração ética constante produzindo descobertas úteis diariamente felizes."
```

```python
# O split transforme a frase ou qualquer texto em uma lista.
palavras = frase.split()
palavras
```

```python
import pandas as pd

tabela = pd.DataFrame({"palavras":palavras})
tabela
```

```python
tabela["proxima_palavra"] = tabela["palavras"].shift(-1)
tabela
```

```python
tabela = tabela.dropna()
tabela
```

```python
filtro = tabela["palavras"] == "Águias"
tabela[filtro]
```

```python
tabela[filtro]["palavras"].value_counts()
```

```python
tabela[filtro]["proxima_palavra"].value_counts(normalize=True)
```

```python
tabela[filtro]["proxima_palavra"].value_counts().idxmax()
```
