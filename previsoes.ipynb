```python
import pandas as pd
import seaborn as sns

df = pd.read_csv(r".\data\base_estudantes_ok.csv")
```

```python
sns.lmplot(data=df, x="study_hours_per_day", y="exam_score");
```

```python
X = df[["age",
   "study_hours_per_day",
   "social_media_hours",
   "netflix_hours",
   "attendance_percentage",
   "sleep_hours",
   "exercise_frequency",
   "mental_health_rating"]]
y = df[["exam_score"]]
```

```python
from sklearn.linear_model import LinearRegression as lr
```

```python
modelo = lr()
modelo.fit(X, y)
```

```python
novo_aluno = pd.DataFrame({
    "age":[18],
   "study_hours_per_day":[1],
   "social_media_hours":[8],
   "netflix_hours":[2],
   "attendance_percentage":[60],
   "sleep_hours":[4],
   "exercise_frequency":[0],
   "mental_health_rating":[1]
    })
modelo.predict(novo_aluno)
```

```python
modelo.score(X,y)
```
