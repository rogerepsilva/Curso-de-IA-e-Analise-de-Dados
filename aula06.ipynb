Gerando uma variável chamada ```colunas```, com todos as colunas que nos traz valores numéricos.

````python
colunas = [
   "age",
   "study_hours_per_day",
   "social_media_hours",
   "netflix_hours",
   "attendance_percentage",
   "sleep_hours",
   "exercise_frequency",
   "mental_health_rating",
   "exam_score"
]
````

Utilizando a biblioteca `Seaborn`, fizemos um mapa de *`heatmap`* correlacionando todas as colunas com **`corr()`**.

Verificamos que em relação a coluna `exam_score` nota do aluno, as outras colunas que estão mais proporcionalmente ligadas são as `study_hours_per_day` e `mental_health_rating`.
Usou-se o comando `fmt=".2f"` para formatar os dados da tabela e não retornar valores grandes.

````python
sns.heatmap(data=df[colunas].corr(), annot=True, cmap="coolwarm", fmt=".2f");
````

Abaixo utilizamos o gráfico ```lmplot```, com o nosso `df` de banco de dados, deixando o **`x`** com a coluna `study_hours_per_day` e **`y`** com a coluna `exam_score`. 
Nos mostrando que a relação entre horas de estudo e a nota no exame está diretamente proporcional.

````python
sns.lmplot(data=df, x="study_hours_per_day", y="exam_score");
````

Vemos abaixo que as horas da netflix, localizadas no **`x`** `netflix_hours`, está inversamente proporcional a nota no exame, localizado em **`y`** `exam_score`.
Onde, quanto mais horas de acesso a netflix, menor a nota.

````python
sns.lmplot(data=df, x="netflix_hours", y="exam_score");
````

Aqui vemos que os estudos por dia `"stydy_hours_per_day"`, não tem nenhuma ligação com a saúde mental `mental_health_rating`. A linha está totalmente na horizontal.

````python
sns.lmplot(data=df, x="study_hours_per_day", y="mental_health_rating")
````

Abaixo criamos duas variáveis, `filtro_horas_altas` para filtrar todas as horas altas (acima de 3 horas de estudo). E `filtro_horas_baixas`, filtrando quem estudou menos que 2 horas.
Realizado a média utilizando `.mean()` e printando essa informação para o usuário.

````python
filtro_horas_altas = df["study_hours_per_day"] >= 3
filtro_horas_baixas = df["study_hours_per_day"] < 2

media_alta = df[filtro_horas_altas]["exam_score"].mean()
media_baixa = df[filtro_horas_baixas]["exam_score"].mean()

print(f"A média das pessoas que estudaram 5 ou mais horas foi {media_alta:.2f}\nJá a média, das pessoas que estudaram menos de duas horas, foi {media_baixa:.2f}")
````
