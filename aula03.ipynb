````python
precos_produtos = [10,5,7,100]
quantidade_produtos = [3,5,6,10]
nomes_produtos = ["Pastel", "Caçulinha", "Refrigerante", "Combos"]

array_quantidade = np.array(quantidade_produtos)
array_produtos = np.array(precos_produtos)

faturamento = array_produtos * array_quantidade

for nome, preco in zip(nomes_produtos, faturamento):
    print(f"{nome}: {preco}")
````

### Exercício de Fixação
- Criar uma lista que irá possuir 10 nomes de produtos.
- Criar uma lista que irá possuir o preço de custo (Atacado).
- Criar uma lista que irá possuir o preço de venda (Varejo).
- Ao final, exibir o lucro obtido em cada produto.

````python
produtos_nome = ["Vinho", "Macarrão","Farofa", "Conhaque", "Refrigerante","Alho","Cerveja", "Café","Estrogonofe","Fini"]
produtos_atac = [139.0, 2.49, 6.89, 5.99, 4.99, 13.90, 4.59, 16.99, 19.19, 7.12]
produtos_var = [169.0, 2.79, 7.89, 6.99, 5.59, 14.90, 6.09, 24.99, 23.99, 9.49]

nomes_array = np.array(produtos_nome) # Não tem necessidade de criar array com essa lista, pois não irá ser utilizado em cálculo.
atacado_array = np.array(produtos_atac)
varejo_array = np.array(produtos_var)

lucro = varejo_array - atacado_array
print(lucro)
print(np.sum(lucro))
print("--" * 40)
i = 0
for nome, custo, venda in zip(nomes_array, atacado_array, varejo_array):
    lucro = venda - custo
    print(f" | {i} | {nome:<12} | teve lucro de {lucro:.2f} |")
    i = i + 1
print("--" * 40)
````
